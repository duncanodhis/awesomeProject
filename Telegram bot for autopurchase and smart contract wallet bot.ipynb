```python
import sqlite3
import threading
from telegram import InlineKeyboardButton, InlineKeyboardMarkup
from telegram.ext import Updater, CommandHandler, CallbackQueryHandler
from telegram import ReplyKeyboardMarkup
from datetime import date

# create a thread-local storage for the database connection
local = threading.local()

def get_db():
    # create a new database connection for each thread
    if not hasattr(local, 'conn'):
        local.conn = sqlite3.connect('auto-shop.db')
    return local.conn

def get_cursor():
    return get_db().cursor()

# create table to store orders
c = get_cursor()
c.execute('''CREATE TABLE IF NOT EXISTS orders
             (id INTEGER PRIMARY KEY AUTOINCREMENT, 
             customer_id INTEGER,
             product TEXT, 
             price REAL,
             date TEXT,
             FOREIGN KEY (customer_id) REFERENCES customers(id))''')
get_db().commit()

# create table to store customers
c = get_cursor()
c.execute('''CREATE TABLE IF NOT EXISTS customers
             (id INTEGER PRIMARY KEY AUTOINCREMENT, 
             customer_id INTEGER UNIQUE,
             city TEXT,
             balance REAL,
             orders INTEGER DEFAULT 0)''')
get_db().commit()

def get_user_data(user_id):
    c = get_cursor()
    c.execute(f"SELECT balance, orders FROM customers WHERE customer_id = {user_id}")
    data = c.fetchone()
    if data:
        return {'balance': data[0], 'orders': data[1]}
    else:
        return {'balance': 0, 'orders': 0}

def start(update, context):
    user_id = update.message.chat_id
    gif_url = 'https://media.giphy.com/media/trN9ht5RlE3Dcwavg2/giphy.gif'
    context.bot.send_animation(chat_id=user_id, animation=gif_url)

    # create a welcome message and inline keyboard with options
    message = 'Welcome to the Auto-Shop Bot! How can we help you today?'
    keyboard = [[InlineKeyboardButton('City', callback_data='city'),
                InlineKeyboardButton('Balance', callback_data='balance')],
                [InlineKeyboardButton('Profile', callback_data='profile'),
                InlineKeyboardButton('Last Order', callback_data='last_order')]]
    reply_markup = InlineKeyboardMarkup(keyboard)

    # send a new message to replace the old one
    message_sent = context.bot.send_message(chat_id=user_id, text=message, reply_markup=reply_markup)

    # edit the message to show only the button that was clicked
    query = update.callback_query
    if query:
        context.bot.edit_message_text(chat_id=user_id, message_id=message_sent.message_id, text=query.data)

def get_products():
    c = get_cursor()
    c.execute("SELECT DISTINCT product FROM orders")
    products = c.fetchall()
    return [product[0] for product in products]

def city(update, context):
    user_id = update.callback_query.message.chat_id
    context.bot.send_photo(chat_id=user_id, photo='https://www.advantour.com/img/georgia/images/tbilisi.jpg')
    message = 'Select an option:'
    
    products = get_products()
    keyboard = [[InlineKeyboardButton(product, callback_data=product)] for product in products]
    keyboard.append([InlineKeyboardButton('Back', callback_data='back')])
    reply_markup = InlineKeyboardMarkup(keyboard)

    context.bot.send_message(chat_id=user_id, text=message, reply_markup=reply_markup)

def balance(update, context):
    user_id = update.callback_query.message.chat_id
    # get user balance from database
    c = get_cursor()
    c.execute(f"SELECT balance FROM customers WHERE customer_id = 3")
    balance = c.fetchone()[0]

    message = f"Your balance: {balance} ლ\nPlease top up your balance:"
    keyboard = [[InlineKeyboardButton("Top Up Balance", callback_data="top_up_balance")]]
    reply_markup = InlineKeyboardMarkup(keyboard)

    context.bot.send_photo(chat_id=user_id, photo='https://www.advantour.com/img/georgia/images/tbilisi.jpg', caption=message, reply_markup=reply_markup)

def profile(update, context):
    user_id = update.callback_query.from_user.id
    c = get_cursor()
    c.execute(f"SELECT city, balance, orders FROM customers WHERE customer_id = {user_id}")
    data = c.fetchone()
    username = update.callback_query.from_user.username or "not written"
    name = update.callback_query.from_user.first_name or "not written"
    surname = update.callback_query.from_user.last_name or "not written"
    balance = data[1] if data else 0
    orders = data[2] if data else 0

    message = f"My Info 👑 Profile👑\n\nUsername: @{username}\nName: {name}\nSurname: {surname}\n\nBalance💷 BALANCE: {balance} GEL\n\nOrders👏 ORDERS: {orders}\n\n"
    keyboard = [[InlineKeyboardButton("Back", callback_data="back")]]
    reply_markup = InlineKeyboardMarkup(keyboard)
    context.bot.edit_message_text(chat_id=user_id, message_id=update.callback_query.message.message_id, text=message, reply_markup=reply_markup)

def last_order(update, context):
    user_id = update.callback_query.from_user.id
    c = get_cursor()
    # get the last order of the user from the orders table
    c.execute(f"SELECT product, price, date FROM orders WHERE customer_id={user_id} ORDER BY date DESC LIMIT 1")
    order_data = c.fetchone()
    if order_data:
        product, price, order_date = order_data
        message = f"Your last order was:\nProduct: {product}\nPrice: {price}\nDate: {order_date}"
    else:
        message = "You have not made any orders yet."
    context.bot.send_message(chat_id=user_id, text=message)

def back(update, context):
    query = update.callback_query
    user_id = query.message.chat_id
    message = 'Welcome to the Auto-Shop Bot! How can we help you today?'
    keyboard = [[InlineKeyboardButton('City', callback_data='city'),
                InlineKeyboardButton('Balance', callback_data='balance')],
                [InlineKeyboardButton('Profile', callback_data='profile'),
                InlineKeyboardButton('Last Order', callback_data='last_order')]]
    reply_markup = InlineKeyboardMarkup(keyboard)
    context.bot.edit_message_text(chat_id=user_id, message_id=query.message.message_id, text=message, reply_markup=reply_markup)


def product_handler(update, context):
    query = update.callback_query
    product = query.data
    user_id = query.message.chat_id
    c = get_cursor()
    c.execute("SELECT price FROM orders WHERE product=?", (product,))
    price = c.fetchone()[0]

    message = f"{product}\nCITY: Tbilisi\nPRICE: {price} l\n\nPlease choose a place:"
    keyboard = [[InlineKeyboardButton("Vazisubani", callback_data="Vazisubani" ),
                 InlineKeyboardButton("TEMQA", callback_data="temqa" )]]
    reply_markup = InlineKeyboardMarkup(keyboard)
    context.bot.edit_message_text(chat_id=user_id, message_id=query.message.message_id, text=message, reply_markup=reply_markup)

def location_handler(update, context):
    query = update.callback_query
    user_id = query.message.chat_id
    _, location, product = query.data.split("_")
    message = f"Chosen location: {location}\nChosen product: {product}\n\nPlease proceed with the purchase or choose another location."
    keyboard = [[InlineKeyboardButton("Purchase", callback_data=f"purchase_{location}_{product}"),
                 InlineKeyboardButton("Back", callback_data="back")]]
    reply_markup = InlineKeyboardMarkup(keyboard)
    context.bot.edit_message_text(chat_id=user_id, message_id=query.message.message_id, text=message, reply_markup=reply_markup)

          
def insert_customer_data(customer_id, city, balance, orders):
    conn = get_db()
    c = conn.cursor()
    try:
        c.execute("INSERT INTO customers (customer_id, city, balance, orders) VALUES (?, ?, ?, ?)",
                  (customer_id, city, balance, orders))
        conn.commit()
        print("Customer data inserted successfully.")
    except Exception as e:
        conn.rollback()
        print("Error inserting customer data:", e)

def insert_order_data(customer_id, product, price, date):
    conn = get_db()
    c = conn.cursor()
    try:
        c.execute("INSERT INTO orders (customer_id, product, price, date) VALUES (?, ?, ?, ?)",
                  (customer_id, product, price, date))
        conn.commit()
        print("Order data inserted successfully.")
    except Exception as e:
        conn.rollback()
        print("Error inserting order data:", e)

def button(update, context):
    query = update.callback_query
    query.answer()
    if query.data == 'city':
        city(update, context)
    elif query.data in get_products():
        product_handler(update, context)
    elif query.data == 'balance':
        balance(update, context)
    elif query.data == 'temqa':
      temqa(update,context)
    elif query.data =='Vazisubani':
      Vazisubani(update,context)
    elif query.data == 'profile':
        profile(update, context)
    elif query.data == 'lastorder':
         last_order(update,context)
    elif query.data == 'lastorder':
         last_order(update,context)
    elif query.data == 'back':
         back(update,context)


  
def main():
   
    updater = Updater('6222839002:AAG3C-gOP-JPQXNHUlkYgZ1EK1j0JBj3w5Y', use_context=True)
    dp = updater.dispatcher
    mydate = date(2023, 3, 31)

    insert_customer_data(3, 'tbilisi', 140, 1)
    insert_customer_data(4, 'tbilisi', 120, 2)
    insert_order_data(1,'1G Amnesia Haze -Super Buds', 140,mydate)
    insert_order_data(2,'10GR Bruce Banner 37% (100 rpm)(800mmm)', 140,mydate)
    insert_order_data(3,'1GR Bruce Banner 27% (100 rpm)(800mmm)', 140,mydate)
          

    # Add handlers
    dp.add_handler(CommandHandler('start', start))
    dp.add_handler(CallbackQueryHandler(city, pattern='^city$'))
    dp.add_handler(CallbackQueryHandler( Vazisubani,pattern='^Vazisubani$'))
    dp.add_handler(CallbackQueryHandler(temqa,pattern ='^temqa$'))
    dp.add_handler(CallbackQueryHandler(balance,pattern ='^balance$' ))
    dp.add_handler(CallbackQueryHandler(profile,pattern ='^profile$' ))
    dp.add_handler(CallbackQueryHandler(last_order, pattern='^last_order$'))
    dp.add_handler(CallbackQueryHandler(back, pattern='^back$'))
    dp.add_handler(CallbackQueryHandler(product_handler))
    
    
    # Start the Bot
    updater.start_polling()
    # Run the bot until you press Ctrl-C
    updater.idle()


if __name__ == '__main__':
    main()

```
